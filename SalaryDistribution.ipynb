```python
%matplotlib nbagg
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from sklearn.preprocessing import scale
import seaborn as sns
```

```python
train  = pd.read_csv('train.csv')
```

```python
salary = (train.Salary).as_matrix()
```

```python
plt.close()
sns.distplot(salary,hist=False,label='probability density of salary')

```

```python
plt.close()
sal = np.log(salary)
sns.distplot(sal,norm_hist=True,hist=True,label='density of log salary')
```
